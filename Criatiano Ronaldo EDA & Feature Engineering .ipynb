```python
#The 'short_basics' and 'range_of_shot' columns are very synomimous and therefore will be similarly 
#converted to numerical columns
#(Less Than 8 ft,Back Court shot, Goal line & Goal_Area can be represented as 1)
#(8-16ft & Penalty Spot can be represented as 2)
#(16-24ft & Mid Range can be represented as 3)
#(24+ft, Right& Left Corner can be represented as 4)
#(Mid Ground Line is represented as 5)
#(nan is represented as 6)
```

```python
data['shot_basics']=data['shot_basics'].map({'Mid Range':3, 'Goal Area':1, 'Goal Line':1, 'Penalty Spot':2, 'nan':6,
       'Right Corner':4, 'Mid Ground Line':5, 'Left Corner':4})
```

```python
data['range_of_shot']=data['range_of_shot'].map({'16-24 ft.':3, '8-16 ft.':2, 'Less Than 8 ft.':1, '24+ ft.':4, 'nan':6,
       'Back Court Shot':1})
```

```python
data['shot_basics']= data['shot_basics'].fillna(data['shot_basics'].mode())
data['range_of_shot']= data['range_of_shot'].fillna(data['range_of_shot'].median())
```

```python
data.isnull().sum()
```

```python
data['area_of_shot'].mode()
```

```python
#Filling the missing values in the #area_of_shot with the modal value
data['area_of_shot']= data['area_of_shot'].fillna(data['area_of_shot'].mode())
data.isnull().sum()
```

```python
# Filling the remaining missing values incase they both have NaN values using the forward fill method
data.shot_basics.fillna(method='ffill', inplace=True)
data.area_of_shot.fillna(method='ffill', inplace=True)
data.range_of_shot.fillna(method='ffill', inplace=True)
data.isnull().sum() # number of missing values for shot_basics and range_of_shot columns should be zero
```

```python
data.head()
```

```python
# Using Forward Filling method in appropriate Columns
print('Null values in column home/away before forward fill =',data['home/away'].isnull().sum())
col = ['home/away','lat/lng', 'team_name']
data.loc[:,col] = data.loc[:,col].ffill()
print('Null values in column home/away after the forward fill =',data['home/away'].isnull().sum())
```
